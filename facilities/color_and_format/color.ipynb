## ✅ Positron 中有效的颜色方案总结

**测试结果：Positron 只支持 LaTeX 数学模式的颜色命令，不支持 HTML 标签。**

### 推荐方案：使用 LaTeX $\color{}$ 命令

#### 基本用法：
*   $\color{red}{红色文字}$
*   $\color{blue}{蓝色文字}$
*   $\color{green}{绿色文字}$
*   $\color{orange}{橙色文字}$
*   $\color{purple}{紫色文字}$

#### 常用颜色名称：
| 颜色名称 | 英文 | 示例 |
|:---|:---|:---|
| 红色 | red | $\color{red}{red}$ |
| 蓝色 | blue | $\color{blue}{blue}$ |
| 绿色 | green | $\color{green}{green}$ |
| 橙色 | orange | $\color{orange}{orange}$ |
| 紫色 | purple | $\color{purple}{purple}$ |
| 青色 | cyan | $\color{cyan}{cyan}$ |
| 洋红 | magenta | $\color{magenta}{magenta}$ |
| 黄色 | yellow | $\color{yellow}{yellow}$ |
| 黑色 | black | $\color{black}{black}$ |
| 白色 | white | $\color{white}{white}$ |
| 灰色 | gray | $\color{gray}{gray}$ |
| 棕色 | brown | $\color{brown}{brown}$ |
| 粉色 | pink | $\color{pink}{pink}$ |
| 深红 | crimson | $\color{crimson}{crimson}$ |

#### 十六进制颜色（RGB）：
*   $\color{#FF0000}{#FF0000 - 红色}$
*   $\color{#3498DB}{#3498DB - 蓝色}$
*   $\color{#2ECC71}{#2ECC71 - 绿色}$
*   $\color{#F39C12}{#F39C12 - 橙色}$
*   $\color{#9B59B6}{#9B59B6 - 紫色}$

#### 混合样式（加重、字体大小）：
*   $\Large{\color{red}{大号红色文字}}$
*   $\small{\color{blue}{小号蓝色文字}}$
*   $\mathbf{\color{green}{加粗绿色文字}}$

### ❌ 以下方法在 Positron 中无效（已测试）

1. **HTML `<span>` 标签**: <span style="color:red">红色</span> (不显示颜色)
2. **HTML `<font>` 标签**: <font color="orange">橙色</font> (不显示颜色)
3. **HTML `<div>` 标签**: <div style="color:blue">蓝色</div> (不显示颜色)
4. **HTML `<p>` 标签**: <p style="color:green">绿色</p> (不显示颜色)
5. **HTML `<mark>` 标签**: <mark>高亮</mark> (不显示颜色)

**原因**: Positron 的 Markdown 渲染器出于安全考虑，过滤了所有 HTML 标签和 inline CSS。
