这是关于工具生态的问题，我直接回答，不涉及代码改动。

先厘清 "cc switch" 到底切换的是什么

cc-switch（CCSwitch）这类工具，本质是帮 Claude Code 切换后端 API 配置（不同的 API key / endpoint / base_url）。它有两种典型用法，结论完全不同，你的两个问题都取决于是哪一种：

1. 换"模型后端"：Claude Code 这个客户端不变，只是把请求指向别的模型（通过 Anthropic 兼容的 API 或中转，比如 GLM、Kimi、DeepSeek，或经 OpenRouter 转的各家模型）。
2. 换"另一家的 agent"：直接不用 Claude Code，改用 OpenAI 的 Codex CLI、Google 的 Gemini CLI 等——这是另一个完全独立的程序，cc-switch 管不到。

---
问题一：Codex 等竞品有没有"月费"选项？

有。主流几家的 CLI agent 都可以走订阅制而非纯 API 按量付费：

┌─────────────┬───────────┬───────────────────────────┬──────────────────────────────────────────────┐
│    Agent    │   所属    │       月费订阅入口        │                     说明                     │
├─────────────┼───────────┼───────────────────────────┼──────────────────────────────────────────────┤
│ Claude Code │ Anthropic │     Claude Pro / Max      │          Pro 约 $20，Max $100/$200           │
├─────────────┼───────────┼───────────────────────────┼──────────────────────────────────────────────┤
│ Codex CLI   │  OpenAI   │    ChatGPT Plus / Pro     │ 含在 ChatGPT 订阅里（Plus 约 $20，Pro $200） │
├─────────────┼───────────┼───────────────────────────┼──────────────────────────────────────────────┤
│ Gemini CLI  │  Google   │ 免费额度 + Google AI 订阅 │                 有较大免费层                 │
└─────────────┴───────────┴───────────────────────────┴──────────────────────────────────────────────┘

所以 Codex 不是只能按 token 付费——你订了 ChatGPT Plus/Pro，Codex CLI 就能用订阅额度跑，这跟 Claude Code 挂在 Claude 订阅上是一样的逻辑。

⚠️ 一个重要区别：订阅额度通常只能配官方那家的模型。如果你用 cc-switch 把 Claude Code 指向第三方模型/中转，走的一般是那家的 API 按量计费或中转商的套餐，不能蹭你的 Claude Max 订阅额度。想用"月费"就得用对应官方客户端 + 官方订阅。

（价格我按截至 2026 年 1 月的情况说，各家调价频繁，下单前请以官网当前页面为准。）

问题二：换了别家模型，原来写的 Claude Code skill 还能用吗？

分两种情况：

情况 1 —— 只换模型后端（仍在 Claude Code 里）：技术上能加载，但可靠性下降。
Skill 本质是 markdown 指令 + Claude Code harness 里的 Skill 工具机制，属于客户端的能力，不属于模型。所以文件照样能被读到、Skill 工具照样存在。但是：
- Skill 的效果依赖模型是否听话地执行指令、是否兼容工具调用格式。非 Claude 模型（尤其是经中转的）常出现：不主动触发 skill、工具调用格式对不上、长指令遵循度差。
- 你那些依赖精细指令的 skill（比如你项目里的 /rnaseq、/wgs、/taxnom）在弱模型上大概率"能调起来但做不对"。

情况 2 —— 换成 Codex / Gemini 等另一个 agent：不能用。
Claude Code 的 skill 是 Claude Code 专有格式，别家 agent 不认。对应关系：
- Codex 用 AGENTS.md
- Gemini CLI 用 GEMINI.md
- 它们各有自己的自定义指令/命令体系，需要你重写一遍。
