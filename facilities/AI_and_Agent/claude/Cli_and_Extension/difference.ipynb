Claude Code 在 CLI（命令行）和 IDE 插件（如 VS Code / Positron Extension）中的工作模式以及功能支持确实是不一样的。

你遇到的 "/plugin isn't available in this environment" 错误，正是一个非常典型的工作模式差异导致的。

### 2. CLI 模式与 App/Extension 模式的核心区别

| 功能维度 | Claude Code CLI 模式 (终端) | IDE Extension 模式 (Positron/VS Code) |
| :--- | :--- | :--- |
| **底层架构** | 原生运行在你的服务器/本地系统，拥有完整的 shell 权限。 | 运行在 IDE 的扩展沙盒内，通过私有包形式托管。 |
| **`/plugin` 安装与生命周期** | **支持。** 可自由安装、更新第三方或官方插件。 | **不支持。** 无法在 UI 窗口通过命令动态增删插件。 |
| **自定义 Slash 命令** | 完美支持通过插件扩展 `/my-command`。 | 仅支持 IDE 预设的基础命令或内置能力。 |
| **环境访问** | 绑定当前终端的 `$PATH`、环境（如 `mamba` 状态）。| 依赖 IDE 暴露的 Workspace 级别 API，相对受限。 |

1. 为什么在 Positron Extension 下会报错？原因：IDE 插件属于“受限/托管环境”在 Claude Code 的架构设计中，/plugin（插件生态）和自适应 Sub-agents（子智能体）的功能，目前仅在独立的 Claude Code CLI（终端命令行）和 Claude Desktop / Cowork 中完整开放。当你在 Positron 或 VS Code 的 UI 扩展插件中使用 Claude Code 时：扩展插件其实是在一个沙盒/托管环境中运行一个私有的 Claude 实例。处于安全性和 IDE 进程稳定性的考虑，IDE 插件的环境不提供底层的文件级、环境级 /plugin 管理工具。因此，在插件的聊天窗口直接输入安装命令，系统会判定当前环境不具备插件挂载能力，返回 isn't available in this environment。
