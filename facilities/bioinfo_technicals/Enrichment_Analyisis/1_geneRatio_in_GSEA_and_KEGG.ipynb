KEGG 和 GSEA 中 的 GeneRatio是什么意思？

在生物信息学分析（特别是使用 clusterProfiler 或 enrichplot 包）中，GeneRatio（基因比例） 是评估通路富集强度的重要指标。

然而，KEGG 传统富集分析（ORA/超几何检验） 和 GSEA（基因集富集分析） 对 GeneRatio 的定义和计算方式有着本质的区别。

🧬 一、 KEGG 经典富集分析（ORA）中的 GeneRatio
在进行常规的差异基因 KEGG 富集分析（如使用 enrichKEGG 函数）时，输出结果中会直接包含 GeneRatio 和 BgRatio 两列。

计算公式：

GeneRatio=  Size/Count  = n/k
​ 
Count (k)：你的差异表达基因（DEGs）列表中，落入该 KEGG 通路的基因数量。

Size (n)：你的差异表达基因列表中，所有能被 KEGG 数据库成功注释的基因总数（注意：不等于全部输入的基因，因为有些基因在 KEGG 中没有对应的通路）。

物理意义：代表在你的差异基因中，有多少比例的基因与该通路相关。

对照组（BgRatio）：背景基因比例（BgRatio= N/M ），即该通路总基因数占整个物种基因库总数的比例。如果 GeneRatio>BgRatio 且 p-value 显著，说明该通路在你的差异基因中被显著富集了。


📊 二、 GSEA 富集分析中的 GeneRatio
如果你去看 GSEA（如 gseKEGG）的原始表格输出，你会发现里面其实没有 GeneRatio 这一列，取而代之的是 ES、NES、Count 和 core_enrichment。

但是，当你用 dotplot() 函数为 GSEA 结果画气泡图时，横坐标默认会变成 GeneRatio。这是绘图包根据 GSEA 的特殊逻辑动态计算出来的。

计算公式：
GeneRatio= setSize/Count
​
Count：指核心富集基因（Core Enrichment Genes 或 Leading Edge Genes）的数量。即在全基因组排序列表中，对该通路富集得分（ES 峰值）做出核心贡献、真正起调控作用的那部分基因。
setSize：该 KEGG 通路本身在数据库中包含的总基因数量（经过设定的最小/最大基因数过滤后）。

物理意义：代表该通路中，有多少比例的核心基因参与了这次整体的协同上调或下调。比例越高，说明该通路被整体激活或抑制得越彻底。

⚖️ 三、 两者核心区别总结
为了帮你彻底分清，我们做个直观的横向对比：

| 特性 / 维度 | KEGG 经典富集 (ORA) 的 GeneRatio | GSEA 富集分析 的 GeneRatio |
| :--- | :--- | :--- |
| **分子（Count）** | 落入该通路的**差异表达基因（DEGs）**数目 | 该通路中做出核心贡献的**核心富集基因（Core Genes）**数目 |
| **分母** | 能被数据库注释的**差异表达基因总数** | **该通路本身**在数据库中所包含的基因总数 |
| **数值含义** | 你的差异基因里有多少比例属于这个通路。 | 这个通路里有多少比例的基因在协同发生变化。 |
| **是否原生存在** | 是，直接存在于结果表格中。 | 否，原始表格中没有，是画 `dotplot` 气泡图时自动计算的。 |
