单菌基因组（即 MAGs）

MAGs 是 Metagenome-Assembled Genomes 的缩写，学术上翻译为“宏基因组组装基因组”。你提到的“单菌基因组”是一个非常形象且准确的通俗称呼。

简单来说，MAGs 就是直接从包含成百上千种微生物的混合环境样本中，通过生物信息学手段“硬生生”剥离、拼接并归类出来的、属于某一种特定微生物的接近完整的基因组。

🧩 一个生动的比喻
想象一下，你手里有 100 套完全不同、大小不一的乐高积木（代表环境中的 100 种细菌）。

打碎测序：你把这 100 套积木全部拆散，混在一个大箱子里，并且把很多长积木断成了碎块（这对应打碎 DNA 进行鸟枪法测序）。

拼接（Assembly）：你把能接得上的碎块先拼成稍微长一点的片段，叫 Contigs（连续外显子片段）。

分箱（Binning）：你虽然没有说明书，但你发现某些积木的塑料材质、颜色深浅非常一致（对应 DNA 的 k-mer 频率），而且在不同的房间里这些积木出现的比例总是同步变化（对应 丰度协同变化模式）。于是你把大箱子里的碎片分堆装进不同的“小盒子（Bins）”里。

最终，每一个被分好堆、能够基本还原出原本面貌的小盒子，就是一个 MAG。 它代表了混合样品中某种特定细菌的基因组蓝图。

🛠️ MAGs 是如何从一锅端的数据中“炼”出来的？
从原始测序数据到得到一个高质量的 MAG，通常要经历三个核心生死关卡：

第一关：组装（Assembly）
把短的 Reads 拼接成较长的 Contigs。由于宏基因组样本里有很多“相似但不同”的菌，这里极其考验算法（如 MEGAHIT 或 SPAdes），一不小心就会把 A 菌的头拼到 B 菌的尾巴上。

第二关：分箱（Binning）
这是形成 MAG 的关键。分箱软件（如 MetaBAT2, SemiBin2）会利用两个核心特征把属于同一只菌的 Contigs 聚类到一起：

核苷酸组成特征：同一物种的基因组往往有特异的四核苷酸频率（Tetranucleotide Frequency, TNF）。

丰度特征：如果在多个样本中，Contig A 变多时 Contig B 也变多，它们大概率来自同一种菌。

第三关：洗白与质控（QC & Refinement）
分出来的“盒子”可能装了脏东西（污染度），也可能丢了零件（完整度）。生物信息学家使用 CheckM 或 CheckM2 利用单拷贝标记基因（SCGs）来评估 MAG 的质量。

📊 MAGs 的质量划分标准（MIMAG 标准）
并不是所有组装出来的 MAG 都能直接用来发文章，国际微生物生态学会（ISME）对其有严格的标准：

高质量 MAG (High-quality Draft)：

完整度（Completeness） > 90%

污染度（Contamination） < 5%

（通常要求包含 5S、16S、23S rRNA 以及至少 18 种 tRNA）

中等质量 MAG (Medium-quality Draft)：

完整度 ≥ 50%

污染度 < 10%

低质量 MAG (Low-quality Draft)：

完整度 < 50% 且 污染度 < 10%（一般不建议用于后续分析）

💡 为什么 MAGs 技术这么火？
在 MAGs 技术成熟之前，我们要想知道一种细菌的基因组，必须先在实验室里把它分离纯化并培养出来（即“数大饼”）。

然而，大自然中超过 99% 的微生物在实验室里是无法人工培养的（被称为“微生物暗物质”）。MAGs 技术直接绕过了“纯化培养”这一世纪难题，让科学家能够直接探知深海、土壤、人体肠道深处那些从未被人类看见过的全新物种的完整基因功能。
